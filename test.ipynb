## Functions

```python
# functions
def gram_schmidt(modes: torch.Tensor) -> torch.Tensor:
    """
    Orthonormalise 4 NxN modes via Gram-Schmidt.
    modes: (B, 4, N, N)  →  returns same shape, orthonormal along dim-1
    """
    B, M, N, _ = modes.shape
    vecs = modes.view(B, M, -1)          # (B, 4, N²)
    out  = []
    for i in range(M):
        v = vecs[:, i]                   # (B, N²)
        for u in out:
            v = v - (v * u).sum(-1, keepdim=True) * u
        v = F.normalize(v, dim=-1)
        out.append(v)
    return torch.stack(out, dim=1).view(B, M, N, N)
```

## Building Blocks

```python
# building blocks
class ConvBlock(nn.Module):
    def __init__(self, in_ch, out_ch):
        super().__init__()
        self.net = nn.Sequential(
            nn.Conv2d(in_ch, out_ch, 3, padding=1, bias=False),
            nn.BatchNorm2d(out_ch),
            nn.LeakyReLU(0.2, inplace=True),
            nn.Conv2d(out_ch, out_ch, 3, padding=1, bias=False),
            nn.BatchNorm2d(out_ch),
            nn.LeakyReLU(0.2, inplace=True),
        )
    def forward(self, x): return self.net(x)

    
```
